```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Organize transformer hyperparameter search results into a single dataframe

```python
def extract_feature(s):
    result = re.search(' Hyper \((.*)\) val loss', s)
    hidden_dim, num_layers, num_heads, dropout, lr, label_weight0, label_weight1 = result.group(1).split(', ')
    # print(int(hidden_dim), int(num_layers), int(num_heads), float(dropout), float(lr), (float(label_weight0[1:]), float(label_weight1[:-1])))

    val_loss = re.search('val loss: (.*) current best val loss', s)
    # print(float(val_loss.group(1)))

    AUPR = re.search('AUPR: (.*) current best AUPR', s)
    # print(float(AUPR.group(1)))
    return int(hidden_dim), int(num_layers), int(num_heads), float(dropout), float(lr), (float(label_weight0[1:]), float(label_weight1[:-1])), float(val_loss.group(1)), float(AUPR.group(1))
```

```python


df = pd.DataFrame(columns=['model', 'hidden_dim',
                  'num_layers', 'num_heads', 'dropout', 'lr', 'label_weight', 'val_loss', 'AUPR'])
with open(r"hyper_files\transformer_all.txt", "r") as f:
    for line in f:
        if line.startswith("==================== Hyper"):
            df.loc[len(df)] = ['transformer'] + list(extract_feature(line))
```

```python
df_sort = df.sort_values(by='AUPR', ascending=False)
df_sort.to_csv(r"hyper_files\transformer_all.csv", index=False)
```

### Organize CNN hyperparameter search results into a single dataframe

```python
def extract_feature_cnn(s):
    result = re.search(' Hyper \((.*)\) val loss', s)
    hidden_dim, kernel_size, dropout, num_layers, lr, label_weight0, label_weight1 = result.group(
        1).split(', ')
    # print(int(hidden_dim), int(num_layers), int(kernel_size), float(
    #     dropout), float(lr), (float(label_weight0[1:]), float(label_weight1[:-1])))

    val_loss = re.search('val loss: (.*) current best val loss', s)
    # print(float(val_loss.group(1)))

    AUPR = re.search('AUPR: (.*) current best AUPR', s)
    # print(float(AUPR.group(1)))
    return int(hidden_dim), int(num_layers), int(kernel_size), float(dropout), float(lr), (float(label_weight0[1:]), float(label_weight1[:-1])), float(val_loss.group(1)), float(AUPR.group(1))
```

```python


df = pd.DataFrame(columns=['model', 'hidden_dim',
                  'num_layers', 'kernel_size', 'dropout', 'lr', 'label_weight', 'val_loss', 'AUPR'])
with open(r"hyper_files\cnn_all.txt", "r") as f:
    for line in f:
        if line.startswith("==================== Hyper"):
            df.loc[len(df)] = ['cnn'] + list(extract_feature_cnn(line))
```

```python
df_sort = df.sort_values(by='AUPR', ascending=False)
df_sort.to_csv(r"hyper_files\cnn_all.csv", index=False)
```

### Transformer

```python
df_transformer.keys()
title_names = ['hidden dimension', 'number of encoder layers',
               'number of attention heads', 'dropout rate', 'learning rate', 'label weight']
```

```python
for i, key in enumerate(df_transformer.keys()[1:-2]):
    a = sns.catplot(
        data=df_transformer, x=key, y="AUPR",
        kind="box", palette="pastel",
    ).set(title=f'Transformer AUPR by {title_names[i]}')
    if i == 5:
        a.set_xticklabels(rotation=0)
```
